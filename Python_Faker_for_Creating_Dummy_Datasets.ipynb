#### Dummy datasets untuk tabel `user`

```python
# Membuat variabel sebagai fungsi Faker() dengan set ('id_ID') untuk Indonesia
fake = Faker('id_ID')

users = []

for i in range(20):
    user = {
        'user_id': i+1,
        'name': fake.name(),
        'phone': fake.phone_number(),
        'address' : fake.address(),
    }

    users.append(user)

df = pd.DataFrame(users)
```

Penjelasan:

1. `fake = Faker('id_ID')`
Membuat objek Faker yang disesuaikan dengan bahasa Indonesia ('id_ID').

2. `users = []`
Membuat sebuah list kosong yang akan digunakan untuk menyimpan data pengguna (users) yang dibuat nantinya.

3. `for i in range(20):`
Loop for yang akan dijalankan sebanyak 20 kali, untuk membuat 20 data.

4. `user = {...}`: Di dalam loop, kita membuat sebuah dictionary yang mewakili data pengguna tunggal.

5. `users.append(user)`: Setiap dictionary pengguna yang dibuat di dalam loop ditambahkan ke dalam list users.

6. `df = pd.DataFrame(users)`: Akhirnya, list users diubah menjadi DataFrame menggunakan pustaka pandas.

```python
# Import Libary yang digunakan
import pandas as pd
from faker import Faker
import faker_commerce
from tabulate import tabulate
import random
from datetime import datetime
```

```python
# Set lokasi Indonesia
fake = Faker('id_ID')

# Menambahkan provifer faker_commerce
!pip install faker-commerce

# Buat variabel banyaknya data
n_cust = 1000
```

### **Start Creating Dummy**

**Table Customer**

```python
# Membuat dictionary kosong table 1
table_customer = {}
table_customer['customer_id'] = [i+1 for i in range(n_cust)]
table_customer['customer_name'] = [fake.name() for i in range(n_cust)]

# Jadikan dictionary menjadi Dataframe
df_table1 = pd.DataFrame(table_customer)

# Tampilkan isi dari dataframe df_table
df_table1
```

**Tabel Customer Details**

```python
# Membuat dictionary kosong table 2
table_customer_detail = {}

table_customer_detail['customer_detail_id'] = table_customer['customer_id']
table_customer_detail['customer_id'] = table_customer['customer_id']
table_customer_detail['address'] = [fake.address() for i in range(n_cust)]
table_customer_detail['postal_code'] = [fake.postcode() for i in range(n_cust)]
table_customer_detail['phone_number'] = [fake.phone_number() for i in range(n_cust)]

# Jadikan dictionary menjadi Dataframe
df_table2 = pd.DataFrame(table_customer_detail)

# Tampilkan isi dari dataframe df_table
df_table2
```

**Table Orders**

```python
def order_date_generator():
    '''
    Fungsi untuk membuat tanggal order

    args:
        None

	return:
		- order_date (str) : string tanggal order
    '''
    # mendefinisikan awal tanggal
    start_date = datetime(2020, 1, 1)

    # mendefinisikan akhir tanggal
    end_date = datetime(2023, 12, 31, 23, 59, 59)

    # membuat dummy berdasarkan start_date dan end_date
    timestamp = fake.date_time_between(start_date = start_date,
                                        end_date = end_date)

    # mengubah objek tanggal menjadi sring
    order_date = timestamp.strftime("%d-%m-%Y %H:%M:%S")

    return order_date
```

```python
# Buat tabel
import random  # Import modul random

table_orders = {}

table_orders["order_id"] = [i+1 for i in range(n_cust)]
table_orders["customer_id"] = [random.choice(table_customer['customer_id']) for i in range(n_cust)]
table_orders["order_date"] = [order_date_generator() for i in range(n_cust)]
table_orders["total_price"] = [fake.random_int(5_000, 120_000, 10_000) for i in range(n_cust)]

# Jadikan dictionary menjadi Dataframe
df_table4 = pd.DataFrame(table_orders)

# Tampilkan isi dari dataframe df_table
df_table4
```

**Table Product**

```python
# Membuat dictionary kosong table 3
table_product = {}
table_product['product_id'] = [f'P-{+ i:003}' for i in range(n_cust)]
table_product['product_name'] = [fake.company() for i in range(n_cust)]
table_product['price'] = [fake.random_int() for i in range(n_cust)]

# Jadikan dictionary menjadi Dataframe
df_table3 = pd.DataFrame(table_product)

# Tampilkan isi dari dataframe df_table
df_table3
```

**Tabel Order Detail**

```python
# Buat table
table_order_detail = {}

table_order_detail["order_id"] = [random.choice(table_orders['order_id']) for i in range(n_cust)]
table_order_detail["product_id"] = [random.choice(table_product['product_id']) for i in range(n_cust)]
table_order_detail["quantity"] = [fake.random_int(1, 10, 1) for i in range(n_cust)]

# Jadikan dictionary menjadi Dataframe
df_table5 = pd.DataFrame(table_order_detail)

# Tampilkan isi dari dataframe df_table
df_table5
```

### **Simpan data ke CSV**

```python
import csv
```

```python
def save_to_csv(data, nama_file):
  # Membuat file csv
  with open(file = f"{nama_file}.csv", mode = 'w', newline = '') as csv_file:
    # Membuat writer csv
    writer = csv.writer(csv_file)

    # write header csv
    writer.writerow(list(data.keys()))

    # mengetahui panjang data
    len_data = len(list(data.items())[0][1])

    # write data ke file csv
    for i in range(len_data):
        row = []
        for key in data.keys():
            row.append(data[key][i])
        writer.writerow(row)
```

    '''
    Fungsi untuk menyimpan data dummy ke csv

    args:
        - data (list)     : list of dictionary data yang akan dijadikan csv
        - nama_file (str) : nama untuk file csv

	return:
		- None
    '''

```python
# save data customer ke file csv
save_to_csv(data = table_customer,
            nama_file = 'table_customer')
```

```python
# save data customer_detail_data ke file csv
save_to_csv(data = table_customer_detail,
            nama_file = 'table_customer_detail')
```

```python
# save data order ke file csv
save_to_csv(data = table_orders,
            nama_file = 'table_orders')
```

```python
# save data produk ke file csv
save_to_csv(data = table_product,
            nama_file = 'table_product')
```

```python
# save data order detail ke file csv
save_to_csv(data = table_order_detail,
            nama_file = 'table_order_detail')
```
